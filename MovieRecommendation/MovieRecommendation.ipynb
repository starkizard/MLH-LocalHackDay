# Feature Extraction and Cleaning

```python
movies = pd.read_csv("movie_dataset.csv")
l = ['keywords','cast','genres','director']
moviesCLEAN = movies[l]
```

```python
movies.head()
```

```python
moviesCLEAN= moviesCLEAN.fillna("")
```

```python
moviesCLEAN.isnull().sum()
```

```python
def combined_feature( dataframe, key, coloumnList ):
    dataframe[key]=""
    for i in coloumnList:
        dataframe[key]+=dataframe[i]+" "
        
combined_feature(moviesCLEAN,"cmb",l)
```

```python
moviesCLEAN["cmb"]
```

# Finding Similarity

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
matrix = CountVectorizer()
out=matrix.fit_transform(moviesCLEAN["cmb"])
```

```python
print(matrix.get_feature_names())
```

```python
print(out.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim = cosine_similarity(out)
sim
```

# Finding the most similar movie

```python
mIndex = 0
while(mIndex < moviesCLEAN.shape[0]):
    if(str(movies["title"][mIndex])=="Avatar"): break
        
```

```python
mIndex
```

```python
movies["title"][0]
```

```python
similarAvatar = sorted(list(enumerate(sim[mIndex])),key= lambda x:x[1] , reverse= True)
```

```python
print("Top 5 movies similar to Avatar are:")
for i in range(1,6):
    print(movies["title"][similarAvatar[i][0]])
```
